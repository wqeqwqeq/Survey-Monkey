```python
def rand_gen(num):
    lst1=list()
    b=np.random.randint(len(final))
    for i in range(num):
        while b not in lst1:
            lst1.append(b)
            continue
        while b in lst1:
            b=np.random.randint(len(final))
            continue
    return lst1
```

```python
final.loc[lst1,'VALUE']=np.nan
```

```python
final['RANGE']=0
```

```python
final.loc[rand_gen(int(len(final)/2)),'RANGE']='[1,1000]'
```

```python
for i in range(len(final)):
    if final.loc[i,'RANGE'] != '[1,1000]':
        final.loc[i,'RANGE'] = '[-100,100]'
```

```python
final.loc[rand_gen(int(len(final)/10)),'RANGE']='Yes/NO'
```

```python
for i in range(len(final)):
    if final.loc[i,'RANGE'] == 'Yes/NO':
        final.loc[i,'RANGE'] = 'Yes/No'
```
